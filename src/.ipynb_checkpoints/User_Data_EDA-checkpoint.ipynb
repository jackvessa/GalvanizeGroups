```python
# Plots
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.set(color_codes=True)
```

## Functions

```python
from datetime import date 
  
def calculateAge(birthDate, today = date.today()): 
    '''
    Function
    --------
    Calculate Age in Years from today
    
    Parameters
    ----------
    birthDate : DateTime Object of Someone's Birthday
    today : Day to calculate Age from
    
    Returns
    -------
    Age in Years
    '''
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day))
  
    return age 
```

```python
from datetime import date 
  
def calculateAge_decimal(birthDate, today = date.today()): 
    '''
    Function
    --------
    Calculate Age in Years, Months, and Days from today (Decimal)
    
    Parameters
    ----------
    birthDate : DateTime Object of Someone's Birthday
    today : Day to calculate Age from
    
    Returns
    -------
    Age (Decimal) - Includes Years, Months, and Days
    '''
    age = today.year - birthDate.year + \
          (today.month - birthDate.month)/12 + \
          (today.day - birthDate.day)/365.25
    
    age = round(age,2)
    
    return age 
```

```python
def set_labels(freq, ax,horz_shift=.2,vert_shift=1000,w="bold",fsize=10,c="b"):
    for i, v in enumerate(freq):
        ax.text(i-horz_shift,v+vert_shift, str(round(v,2)), weight = w,fontsize=fsize,color = c)
```

## User Information

```python
user_donations_df = pd.read_csv('../data/User_Donations_File_Acct.csv')
```

```python
user_donations_df.head()
```

```python
user_donations_df.groupby('Gender').size()
```

### Birth Information

```python
user_donations_df['Birth_Year'] = user_donations_df['Birthday'].apply(lambda x: int(pd.to_datetime(x).year) if
                                                                     not math.isnan(x.year) else x)
```

```python
user_donations_df['Birthday'] = user_donations_df['Birthday'].apply(
    lambda x: pd.to_datetime(x))
```

```python
user_donations_df['User_Age'] = user_donations_df['Birthday'].apply(lambda x : 
                    calculateAge_decimal(date(int(x.year),int(x.month),int(x.day))) 
                    if not math.isnan(x.year) else x)
```

### Date Created Information

```python
user_donations_df['Date created'] = user_donations_df['Date created'].apply(
    lambda x: pd.to_datetime(x))
```

```python
user_donations_df['Date_Created_Year'] = user_donations_df['Date created'].apply(lambda x: pd.to_datetime(x).year)
```

```python
user_donations_df['Date_Created_Month'] = user_donations_df['Date created'].apply(lambda x: pd.to_datetime(x).month)
```

```python
user_donations_df['Account_Age'] = user_donations_df['Date created'].apply(lambda x : 
                    calculateAge_decimal(date(int(x.year),int(x.month),int(x.day))) 
                    if not math.isnan(x.year) else x)
```

```python
user_donations_df.head()
```

## EDA + Graphs

### Age of Account Information

```python
year_freq = user_donations_df.groupby('Date_Created_Year').size()
years = [2017,2018,2019]

fig, ax = plt.subplots(figsize=(12,8))
ax = sns.barplot(x=years,y=year_freq,color='b')

# Title
ax.set_title(label="Account Signups by Years",weight='bold',fontsize=30)

# X-axis
ax.set_xticklabels(labels=years,weight='bold',fontsize=25,rotation=0)
ax.set_xlabel(xlabel="Years",weight='bold',fontsize=30)

# Y-axis
y_tick_nums = np.arange(0,1601,200)
ax.set_yticks(y_tick_nums)
ax.set_yticklabels(labels=y_tick_nums,weight='bold',fontsize=20)
ax.set_ylabel(ylabel="Frequency",weight='bold',fontsize=25)

# Frequency Labels
set_labels(year_freq, ax,horz_shift=0.10,vert_shift=-100,w="bold",fsize=25,c="orange")

plt.tight_layout()
```

```python
month_freq = user_donations_df.groupby('Date_Created_Month').size()
months = ['January','February','March','April','May','June',
          'July','August','September','October','November','December']

fig, ax = plt.subplots(figsize=(12,8))
ax = sns.barplot(x=months,y=month_freq,color='b')

# Title
ax.set_title(label="Account Signups by Months",weight='bold',fontsize=30)

# X-axis
ax.set_xticklabels(labels=months,weight='bold',fontsize=20,rotation=35)
ax.set_xlabel(xlabel="Months",weight='bold',fontsize=25)

# Y-axis
y_tick_nums = np.arange(0,401,50)
ax.set_yticks(y_tick_nums)
ax.set_yticklabels(labels=y_tick_nums,weight='bold',fontsize=20)
ax.set_ylabel(ylabel="Frequency",weight='bold',fontsize=25)

# Frequency Labels
set_labels(month_freq, ax,horz_shift=.25,vert_shift=-20,w="bold",fsize=20,c="orange")

plt.tight_layout()
```
