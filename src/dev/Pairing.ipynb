```python
import numpy as np
import pandas as pd
```

```python
past_df = pd.read_excel('./past_pairs.xlsx')
```

```python
past_df.fillna('',inplace=True)
past_df
```

```python
new_df = pd.read_excel('./new_pairs.xlsx')
```

```python
new_df.fillna('',inplace=True)
new_df.head()
```

```python
def create_pairs(df):
    pairs = set()
    trios = set()
    xPairs = set()
    for i,j,k in zip(df['pair1'], df['pair2'], df['pair3']):
        if i == '':
            continue
        if k != '':
            trios.add(frozenset([i,j,k]))
            xPairs.add(frozenset([i,j]))
            xPairs.add(frozenset([i,k]))
            xPairs.add(frozenset([k,j]))
        else:
            pairs.add(frozenset([i,j]))
    return pairs, trios, xPairs

past_pairs, past_trios, past_xPairs = create_pairs(past_df)
new_pairs, new_trios, new_xPairs = create_pairs(new_df)
past_xPairs
```

```python
matches = past_pairs & new_pairs
if len(matches) > 0:
    print('Pairs repeat:', matches)
else:
    print('No exact pair repeat found')
```

```python
matches = past_trios & new_trios
if len(matches) > 0:
    print('Trios repeat:', matches)
else:
    print('No trio repeat found')
```

```python
matches = (past_xPairs & new_pairs) | (past_pairs & new_xPairs)
if len(matches) > 0:
    print('A pair was found in a trio:', matches)
else:
    print('No pair in trio repeat found')
```
